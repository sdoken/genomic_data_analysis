```python
import pandas as pd
import plotly
import plotly.graph_objs as go
```

```python
raw_data = pd.read_csv('./CCLE.rpkm.v2.gct', sep='\t', skiprows=2)
data = raw_data.set_index('Description').drop('Name', axis=1).T
```

```python
radsen_table_raw = pd.read_csv('./RadSen_CTD2_new.xlsx', sep='\t', skiprows=2)
radsen_table = radsen_table_raw.set_index('Label').drop('Description', axis=1).T
```

```python
joined_data = pd.concat([data, radsen_table[['AUC', 'Normalized_AUC']]], axis=1, sort=False)
joined_data.dropna(axis=0, how='all', subset=['AUC', 'Normalized_AUC'], inplace=True)
```

```python
def plot_gene(gene_name):
    x_values = list(joined_data[gene_name])
    y_values = list(joined_data['AUC'])
    plotly.offline.init_notebook_mode(connected=True)
    plotly.offline.iplot({ "data": [go.Scatter(x=x_values, y=y_values, mode = 'markers', text = list(joined_data.index))],
    "layout": go.Layout(title=gene_name)
})
```

#### The cell below is where you would enter gene name in quotes and it would plot the data points for you:

```python
plot_gene('FAM138A')
```
